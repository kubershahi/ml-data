### Importing the necessary package

```python
import pandas as pd
import requests
```

```python
# reading the dataset
df = pd.read_csv('article.csv').iloc[:,1:]
df
```

```python
#using the huggingface api inference to summarize the sections using sshleifer/distilbart-cnn-6-6

API_URL = "https://api-inference.huggingface.co/models/sshleifer/distilbart-cnn-6-6"
headers = {"Authorization": f"Bearer {API Token}"} # has my hugging face token: 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

# function for POST Reequest and GET the response
def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

# get the summary of each section in the dataframe['Original Text']
output = query({
    "inputs": df['Original Text'].to_list(),
})
```

```python
# create a list of response summaries
summary = []

for i in range(len(output)):
    summary.append(output[i]['summary_text'])

print(len(summary))
print(summary)
```

```python
# add summary to the original dataframe

df['Summary'] = summary
df
```

```python
# save the dataframe as csv and json file

df.to_csv('article_summaries.csv')
df.to_json('article_summaries.json')
```
